```python
import matplotlib.pyplot as plt
import seaborn as sns

import generate_data
import numpy as np

import pylab 
```

```python
fxnames = ["USDJPY", "USDCHF", "USDCAD", "NZDUSD", "GBPUSD", "EURUSD", "AUDUSD"]

prefix = "../data/forex_majors_candle/"
suffix = "-2016_01_01-2017_06_27.csv"
filenames = [prefix+m+suffix for m in fxnames]

data = {}
times = {}
hours = {}
count = 0
for filename in filenames:
    tempdata, time, hour, dmy = generate_data.readduka(filename)

    # Invert if USD comes first
    if fxnames[count][:3] == "USD": 
        tempdata = 1.0/tempdata
        temp = tempdata[:, 2]
        tempdata[:, 2] = tempdata[:, 3]
        tempdata[:, 3] = temp

    data[fxnames[count]] = tempdata
    times[fxnames[count]] = time
    hours[fxnames[count]] = hour
    count+=1
```

```python
removethese = generate_data.get_uncommon_index_from_ordered_list_of_lists([times[fn] for fn in fxnames])

for i in range(len(fxnames)):
    mask = np.ones(data[fxnames[i]].shape[0]).astype(np.bool_)
    mask[removethese[i]] = False
    data[fxnames[i]] = data[fxnames[i]][mask]
    times[fxnames[i]] = times[fxnames[i]][mask]
    hours[fxnames[i]] = hours[fxnames[i]][mask]
```
